```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

# from catboost import CatBoostRegressor, Pool
```

```python
all_data = pd.read_csv('./data/final_2022_125.csv')
```

```python
all_data.head(1)
```

```python
all_data['calls_per_day'] = all_data.groupby(['dt'])['id'].transform('count')
```

```python
all_data[all_data.dt > '2020-02'].groupby('calls_per_day')[['dt','calls_per_day']].first().sort_values(by='dt').set_index('dt').plot(figsize=(16,6), title='Вызовы по дням')
```
